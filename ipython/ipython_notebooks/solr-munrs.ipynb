```python
import json
import pysolr
import pandas as pd
import numpy as np

from pandas import DataFrame

SOLR_URL = "localhost:8983/solr/munrs"
si = pysolr.Solr('http://localhost:8983/solr/munrs', timeout=10)
```

```python
result = si.search("*:*",rows=800,fl="area_tf,codigo_ibge_s,dens_pop_tf,gentilico_s,nome_s,pib_ti,pop2010_ti")
```

```python
df = DataFrame.from_records(result.docs,columns=['nome_s','pib_ti','pop2010_ti','dens_pop_tf'])
df.columns = ['nome', 'pib','pop2010','dens_pop']

df.head(20)

```

```python
pib = df['pib']
print pib.sum()
print pib.count()
print pib.max()
```

```python
def get_rank(field):
    print "Campo=", field
    rank = df[field].rank(ascending=True, method='max')
    print "idxmin - Indice dos valor minimo"
    imin = rank.idxmin()
    imax = rank.idxmax()
    print "imin=",imin
    print df.ix[imin]
    print "-"*20
    print "imax=",imax
    print df.ix[imax]
```
